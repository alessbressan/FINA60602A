```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('data/48_Industry_Portfolios_Daily.csv', skiprows=9, nrows= 25880, skipinitialspace=True,)
df.rename(columns={'Unnamed: 0' : 'Date'}, inplace= True)
df.columns = df.columns.str.strip()
df.loc[:, df.columns != 'Date'] = df.loc[:, df.columns != 'Date'].apply(pd.to_numeric, errors= 'coerce')
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d').dt.strftime('%Y-%m-%d')
df.to_csv('clean_data/daily_data.csv')
df.tail()
```
